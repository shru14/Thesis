```python
import os
import json

# Define input file and output directory paths
input_file = r"C:\Users\91909\Desktop\case-law\A\modified_final_opinions.json"
output_dir = r"C:\Users\91909\Desktop\case-law\A\chunks\big"

# Create the output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Load the input JSON file (assumed to be a list of records)
with open(input_file, "r", encoding="utf-8") as f:
    data = json.load(f)

# Set the chunk size
chunk_size = 5000

# Get the total number of records
total_records = len(data)

# Iterate over the data in steps of chunk_size
chunk_index = 1
for start in range(0, total_records, chunk_size):
    end = start + chunk_size
    chunk_data = data[start:end]
    
    # Define the file path for the current chunk
    chunk_filename = os.path.join(output_dir, f"chunk_{chunk_index}.json")
    
    # Write the current chunk to the JSON file
    with open(chunk_filename, "w", encoding="utf-8") as outfile:
        json.dump(chunk_data, outfile, indent=2)
    
    print(f"Chunk {chunk_index} saved: {len(chunk_data)} records (indexes {start} to {end - 1}).")
    chunk_index += 1

print(f"\nDividing complete. All chunks (each up to {chunk_size} records) are saved in {output_dir}.")
```
